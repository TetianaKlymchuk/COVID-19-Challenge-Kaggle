 Inicialization of a class. Can be done in 2 ways:
 
     1) providing a path to saved model
 
     2) providing an inicial dataframe
 
In the first case saved model will be loaded
In the second case model will be trained from zero using provided df

```python
m = Embedding_doc2vec(path='doc2vecmodel.mod')
```

Chosing one paper_id for testing

Getting all the distances to a given paper_id

```python
distances = m.get_distances(df, paper_id)
```

```python
import scipy
import re
import gensim
text_1 = df[df.paper_id == paper_id].text.values[0]
text_2 = gensim.utils.simple_preprocess(text_1)
vector_1 = m.get_embedding(paper_id, df)
vector_2 = m.model.infer_vector(doc_words=text_2, steps=20, alpha=0.025)
dist = scipy.spatial.distance.cosine(vector_1, vector_2)
print(dist)
```
